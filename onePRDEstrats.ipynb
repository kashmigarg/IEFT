```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

timeall = []
strat = []
bprof = []
bstrat = []
def pps(k,f):
    data = pd.read_csv('1PRDEK'+str(k)+'F'+str(f)+'_ZIC201_strats.csv', header=None)
    row = 0
    while row < len(data):
        timeall.append(data.iloc[row, 1])
        strat.append(data.iloc[row, 6])
#         prof.append(data.iloc[row, 8])
        bprof.append(data.iloc[row, 12])
        bstrat.append(data.iloc[row, 14])
        row += 1
#     print(strat)
    fig, bx = plt.subplots()
    fig, ax = plt.subplots()
    fig, cx = plt.subplots()
    bx.plot(bstrat, bprof, '*', color='black')
    ax.plot(timeall, bstrat, 'b-', color='black')
    cx.plot(timeall, bprof, 'x', color='black')
    plt.plot()

pps(4,0.5)
# pps(10,0.8)
# pps(10,1.5)
```
