```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

days_avg = 7
# total_days = 100

def pps(file):
    timeall = []
    bbstrat = []
    bsstrat = []
    data = pd.read_csv(file, header=None)
    row = 0
    while row < len(data):
        timeall.append(data.iloc[row, 1])
        bbstrat.append(data.iloc[row, 217])
        bsstrat.append(data.iloc[row, 223])
        row += 1
    return timeall, bbstrat, bsstrat
    
    
timeall, bbstrat, bsstrat = pps('15PRDE14K.8F1_strats.csv')
fig, ax = plt.subplots()
fig, bx = plt.subplots()
ax.plot(timeall, bbstrat, 'x', color='black')
bx.plot(timeall, bsstrat, 'x', color='red')

timeall, bbstrat, bsstrat = pps('15PRDE14K1.4F1_strats.csv')
fig, ax = plt.subplots()
fig, bx = plt.subplots()
ax.plot(timeall, bbstrat, 'x', color='black')
bx.plot(timeall, bsstrat, 'x', color='red')

timeall, bbstrat, bsstrat = pps('15PRDE14K2F1_strats.csv')
fig, ax = plt.subplots()
fig, bx = plt.subplots()
ax.plot(timeall, bbstrat, 'x', color='black')
bx.plot(timeall, bsstrat, 'x', color='red')

# timeall, bbstrat, bsstrat = pps('15PRDE9K2F1_strats.csv')
# fig, ax = plt.subplots()
# fig, bx = plt.subplots()
# ax.plot(timeall, bbstrat, 'x', color='black')
# bx.plot(timeall, bsstrat, 'x', color='red')

plt.plot()
```
